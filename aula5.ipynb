```python
import seaborn as sns
```

```python
# Como definir a coluna Pais como indica
df.set_index('País',inplace=True)
```

```python
df
```

```python
top_10 = df.sort_values(by='Total',ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
ax = sns.barplot(data=top_10,y=top_10.index,x='Total',orient='h')
ax.set(
    title='Os dez países com mais imigrantes para o Canadá',
    xlabel='Total de imigrantes',
    ylabel=''

)

plt.show()
```

```python
def gerar_grafico_paleta('pallete'):
  fig,ax =plt.subplots(figsize=8,4)
  ax = sns.barplot(data=top_10,y=top_10.index,x='Total',
                   oritent='h',
                   palete=palette)

  sns.color_palette("dark:#5A9_r", as_cmap=True)
  ax.set_title('Os dez países com mais imigrantes para o Canadá',
              loc='left',
              fontsize=16)
  ax.set_xlabel('Ttoal de imigrantes',fontsize=14)
  ax.set_ylabel('')
  plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10,y=top_10.index,x='Total',orient='h', palette='Spectral')

#sns.color_palette("dark:#5A9_r", as_cmap=True)
ax.set_title('Os dez países com mais imigrantes para o Canadá',
              loc='left',
              fontsize=16)
ax.set_xlabel('Ttoal de imigrantes',fontsize=14)
ax.set_ylabel('')
plt.show()
```

# Graficos Interativos com Plotly

```python
import plotly.express as px
```

```python
anos = list(map(str,range(1980,2014)))
```

```python
brasil = df.loc['Brasil',anos]
```

```python
brasil_dict = {'ano':brasil.index.tolist(),'imigrantes':brasil.values.tolist()}
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil.head()
```

```python
fig = px.line(dados_brasil,x='ano',y='imigrantes',title='imigrantes no Brasil')
fig.show()
```
