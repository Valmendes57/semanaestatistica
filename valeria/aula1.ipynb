- Temos 195 strings
- Vamos pegar a coluna país.

```python
# usaremos uma função chamada setindex
df['País']
# seleciona a coluna, sem sinal de igual, tá apenas selecionando.
```

```python
# setar como índice
df.set_index('País',inplace=True)
# interessante pois isso faz com que eu possa selecionar os índices como quiser
```

```python
anos=list(map(str, range(1980,2014)))
# O map ele quer saber o que fazer
# colocando uma classe lista, para pegar uma data de 2000 a 2013 por exemplo
# Lembrando que ele transformou os int em string e percorreu da maneira especificada no parâmetro, de 1980 a 2014
```

```python
anos
```

```python
# A funcao loc pega o país como índice
brasil=df.loc['Brasil', anos]
```

```python
brasil
```

```python
print(type(brasil))
```

```python
# tipos de dados que o python e pandas possuem
# O pandas possui os tipos= Series e DataFrames
# Os tipos do Python são= Lista e Tupla e Dict
# O Dicionário é uma espécie de tabela, tem valores.
# Dicionario possuem chaves e valores.
# O arroz: chave, Valor é uma descrição
brasil_dict={'ano': brasil.index.tolist(), 'imigrantes':brasil.values.tolist()}
# Uma series, é um vetor de uma dimensão, nao é uma tabela, mas estou transformando a series em Dicionario
# Ano virou index, por ser chave
# Imigrantes são os valores
# Dicionário se inicia com as chaves
# vamos passar nas chaves, a chave e o valor
```

```python
brasil_dict
```

```python
dados_brasil=pd.DataFrame(brasil_dict)
```

```python
dados_brasil
# todo esse processo se chama Transformação dos Dados.
```

```python
! pip install matplotlib
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'])
# Quando chamo o plot e passo os colchetes, estou selecionando as colunas do eixo x e y
# é interessante as alterações todas antes do plt.show()
# Manipulando o eixo X
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
# Manipulando o eixo Y
plt.yticks([500,1000,1500,2500,3000]) # Estes são os valores, sem aspas
plt.show()
# Ficou muito bagunçado, então vamos arrumar la embaixo
```

# Transformando gráficos em figura

```python
plt.figure(figsize=(8,4))
# 1 polegada = 2.54 cm
# largura 8 = 20.32 cm
# altura 4 = 10.16 cm
# DPI ( dots per inch, ou pontos por polegadas)
# PAdrão no matplotlib - DPI = 100
# LArgura : 8 polegadas x 100 DPI = 800 pixels.
# Altura : 4 polegadas x 100 DPI = 400 pixels.
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.yticks([500,1000,1500,2500,3000])
plt.grid()
plt.title("Imigração do Brasil para o Canadá")
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.show()
```

# Inicio da aula 2

```python
# PRecisarei de duas variáveis para vários gráficos?
# subplot ( criando a figura )
fig,axis=plt.subplots(1,2,figsize=(15,6))
# fig e ax sao variaveis que armazenam o plt.subplots
axis[0].plot(dados_brasil['ano'],dados_brasil['imigrantes'],color="brown")
axis[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axis[0].set_title('Migração do Brasil para o canadá de 1980-2013')
axis[0].set_xlabel('Ano')
axis[0].set_ylabel('Numero de imigrantes')
axis[0].grid()
# Agora o index 1....
axis[1].boxplot(dados_brasil['imigrantes'])
axis[1].set_title("Boxplot da imigração do Brasil para o Canadá")
axis[1].set_xlabel('brasil')
axis[1].grid()
axis[1].set_ylabel('Numero de imigrantes')
plt.show()
```

# Comparativo de imigrações do Brasil com alguns países da América Latina

```python
# Criar a figura  
fig,axs=plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.9,wspace=0.5)
# O loop for usa a função flat para "achatar" a matriz bidimensional em uma única sequencia de eixos.
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
for ax in axs.flat:
    ax.set_xlabel('Ano', fontsize=10)
    ax.set_ylabel('Numeros de imigrantes', fontsize=10)

# definindo os limites minimos e maximos dos valores do eixo y, pode ser feito com o eixo x também
ymin=0
ymax=7000
for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)

axs[0,0].plot(df.loc['Brasil',anos],color='green')
axs[0,0].set_title('Brasil')

# Indexado do Brasil
axs[0,1].plot(df.loc['Colômbia',anos],color='orange')
axs[0,1].set_title('Colômbia')

# Colombia
axs[1,0].plot(df.loc['Argentina',anos],color='blue')
axs[1,0].set_title('Argentina')

# Argentina
axs[1,1].plot(df.loc['Peru',anos],color='red')
axs[1,1].set_title('Peru')
```
